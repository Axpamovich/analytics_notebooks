### Как часто люди возвращаются?

Используем метрику Retention Rate для когорт по месяцам.

```python
# время первого посещения
first_visit_date = visits.groupby('uid')['start_ts'].min()
first_visit_date.name = 'first_visit_date'
visits = visits.join(first_visit_date, on = 'uid')
visits['first_month'] = visits['first_visit_date'].astype('datetime64[M]')

# расчет lifetime когорты
visits['cohort_lifetime'] = visits['month'] - visits['first_month']
visits['cohort_lifetime'] = visits['cohort_lifetime']/np.timedelta64(1, 'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype(int)
```

```python
# сгруппированные по когортам данные
cohorts = visits.groupby(['first_month',
                          'cohort_lifetime']).agg({'uid':'nunique'}).reset_index()

# изначальное количество пользователей в когорте
inital_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_month','uid']]

inital_users_count = inital_users_count.rename(columns={'uid':'cohort_users'})
cohorts = cohorts.merge(inital_users_count,on='first_month')

# расчет retention rate
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users']

# сводная таблица
retention_pivot = cohorts.pivot_table(
    index='first_month',
    columns='cohort_lifetime',
    values='retention', aggfunc='sum')
```

```python
# тепловая карта retention rate по месяцам
plt.figure(figsize=(14, 9))
plt.title('retention rate by months')
ax = sns.heatmap(retention_pivot, annot=True, fmt='.2%', linewidths=0.5, linecolor='gray')
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

Это довольно низкий, но вполне объяснимый показатель того, как люди возвращаются. 

Предположительно, дело в том, что абсолютно все равно, на каком сервисе покупать билеты - везде примерно одинаковая комиссия и примерно одинаковые условия, и люди используют тот сайт, который попался под руку. 

Это, кстати, подтверждается другим показателем: среднее количество сессий в день (984) совсем незначительно превышает среднее количество уникальных пользовтелей в день (907), что говорит о том, что почти все пользователи - уникальные, появляются один раз и возвращаются редко.

## Метрики продаж <a name="2.2"></a>

### Когда люди начинают покупать?

Найдем среднее время разницы между первым визитом и первой покупкой: время первого визита у нас уже есть, осталось определить время первого заказа.

```python
# к таблице про заказы добавим столбец со временем первого посещения
orders = orders.join(first_visit_date, on = 'uid')
# проверим, есть ли данные о заказах, где время первого посещения больше времени заказа
len(orders[orders['buy_ts'] < orders ['first_visit_date']])
```

```python
# посмотрим, как это выглядит
orders[orders['buy_ts'] < orders ['first_visit_date']].head(20)
```

Данных, где время первого посещения больше времени заказа, сравнительно много, и закономерность ничем не объясняется: если бы, например, время в двух этих выгрузках учитывалось бы для разных часовых поясов, то проблем было бы больше.

Думаю, что можно их выбросить, их чуть больше одного процента.


```python
# выбрасываем заказы, где время заказа меньше времени первого посещения
orders = orders[orders['buy_ts'] >= orders ['first_visit_date']]
orders.info()
```

Чтобы узнать, как меняется количество заказов в зависимости от рекламного источника и устройства, надо добавить эти данные в таблицу заказов. 

```python
# сделаем срез - оставим только первые визиты, чтобы только информацию по ним добавлять в таблицу orders
first_visits = visits[visits['first_visit_date'] == visits['start_ts']]
first_visits.info()
```

```python
orders = orders.merge(first_visits[['uid', 'device', 'source_id']], on = ['uid'], how = 'left')
orders = orders[orders['source_id']!=7]
orders.info()
```

```python
# время первого заказа
first_order_date = orders.groupby('uid')['buy_ts'].min()
first_order_date.name = 'first_order_date'
orders = orders.join(first_order_date, on = 'uid')

# разница между временем первого посещения и временем первого заказа
orders['time_lag'] = (orders['first_order_date'] - orders['first_visit_date']).dt.seconds
# распределение значений
orders.hist('time_lag', bins = 50)
plt.show()
```

```python
orders['time_lag'].mode()
```

Это то же время, что у нас получилось как самое часто встречающееся значение для продолжительности сессии.

Надо полагать, этого хватает, чтобы сделать заказ; можно сделать вывод о том, что люди приходят и покупают сразу, в момент посещения.

### Сколько раз покупают за период?

```python
orders_per_day = orders.groupby('day').agg({'uid': 'count'}).mean()
orders_per_month = orders.groupby('month').agg({'uid': 'count'}).mean()
orders_per_week = orders.groupby('week').agg({'uid': 'count'}).mean()

print('Среднее количество покупок в день -', int(orders_per_day))
print('Среднее количество покупок в неделю -', int(orders_per_week))
print('Среднее количество покупок в месяц -', int(orders_per_month))
```

Получим также данные по количеству заказов для когорт, чтобы увидеть более подробную картину. Когорты будем назначать по дате первого визита - мы уже видели, что она в основном совпадает с датой первого заказа.

```python
# месяц первого визита
orders['first_month'] = orders['first_visit_date'].astype('datetime64[M]')

# расчет lifetime когорты
orders['cohort_lifetime'] = orders['month'] - orders['first_month']
orders['cohort_lifetime'] = orders['cohort_lifetime']/np.timedelta64(1, 'M')
orders['cohort_lifetime'] = orders['cohort_lifetime'].round().astype(int)

# сгруппированные по когортам данные по заказам, общей выручке и уникальными пользователями
cohorts_orders = orders.groupby(['first_month',
                          'cohort_lifetime']).agg({'revenue': 'sum', 'uid':['count', 'nunique']}).reset_index()
# новые названия колонок
cohorts_orders.set_axis(['first_month', 'cohort_lifetime', 'revenue', 'orders_count', 'n_users'], axis = 1, inplace = True)

# сводная таблица
orders_pivot = cohorts_orders.pivot_table(index='first_month',columns='cohort_lifetime',values='orders_count', aggfunc='sum')
```

```python
# тепловая карта с количеством заказов по месяцам
plt.figure(figsize=(14, 9))
plt.title('количество заказов по месяцам')
ax = sns.heatmap(orders_pivot, annot=True,cmap='Blues', fmt='g', linewidths=0.5, linecolor='gray')
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

Количество заказов для когортного анализа падает примерно так же, как и в когортном анализе для посещаемости, и виден тот же пик для осени и зимы, что и везде.

```python
# как количество заказов меняется по месяцам в зависимости от рекламного источника
orders_source = orders.pivot_table(index = 'month', columns = 'source_id', values = 'uid', aggfunc = 'count')
orders_source.plot(figsize = (13,7), grid = True)
plt.title('Количество заказов по рекламным источникам в месяц')
plt.show()
```

```python
# как количество заказов меняется по месяцам в зависимости от устройства
orders_device = orders.pivot_table(index = 'month', columns = 'device', values = 'uid', aggfunc = 'count')

orders_device.plot(figsize = (13,7), grid = True)
plt.title('Количество заказов по устройствам в месяц')
plt.show()
```

Распределение заказов совпадает с распределением [посещений](#up) по рекламным источникам в месяц - самые активные источники это 3, 4 и 5. Картина с количеством заказов по устройствам также похожа на количество визитов по устройствам, причем разница между количеством покупок с компьютера и со смартфона даже выраженнее, чем разница между количеством посещений. Это означает, что если просто изучать афишу с телефона еще можно, хоть и неудобно, то покупать неудобно совсем.

### Сколько денег приносят? (LTV)

У нас уже есть количество уникальных пользователей и выручка на каждый отдельный месяц по когортам, осталось только свести их вместе на всю продолжительность существования когорт.

```python
# выручка с пользователя за каждый месяц по когортам
cohorts_orders['revenue_per_user'] = cohorts_orders['revenue']/cohorts_orders['n_users']

# выручка с пользователя за все время существования когорты
report_ltv = cohorts_orders[(cohorts_orders['first_month']<='2017-12-01')&(cohorts_orders['cohort_lifetime']<=6)]['revenue_per_user'].mean()

print ('Средние траты пользователя за полгода:', report_ltv)
```
